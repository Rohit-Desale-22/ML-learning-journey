```python
name=[]
ratings=[]
reviews=[]
for j in range(1,10): 
    webpage = requests.get('https://www.ambitionbox.com/list-of-companies?campaign=desktop_nav&page={}'.format(j),headers=headers).text
    soup = BeautifulSoup(webpage, 'html.parser')
    company = soup.find_all('div', class_ = 'companyCardWrapper')
    for i in company:
        name.append(i.find('h2').text.strip())
        ratings.append(i.find('div', class_='rating_text rating_text--md').text.strip())
        for tag in i.find_all('a', class_ = 'companyCardWrapper__ActionWrapper'):
            href = tag.get('href', '')
            if 'reviews' in href:
                reviews.append(i.find('span', class_='companyCardWrapper__ActionCount').text.strip())
```

```python
df=pd.DataFrame({'company name': name, 'ratings': ratings, 'reviews': reviews})
```
